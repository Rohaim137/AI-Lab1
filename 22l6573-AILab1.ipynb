```python
#Function: Factorial Calculator
def factorial(n):
  facto = 1
  for i in range(n, 0,-1):
    facto = facto * i
  return facto
number= int(input("Enter number: "))
print(factorial(number))
```

```python
#Average Calculator with Validation
try:
  numbers = input("Enter numbers with spaces: ")
  listOfNums = numbers.split()
  sum = 0
  for i in range(0,len(listOfNums)):
    listOfNums[i] = int(listOfNums[i])
    sum = sum + listOfNums[i]
  print(sum/len(listOfNums))
except:
  print("Invalid input")
```

```python
#User Registration System
user_info = {}
def signUp(username, password):
  if username in user_info:
    print("Username already exists")
    return False
  else:
    user_info[username] = password
    print("Registration Successful")
    return True
def login(username, password):
  if username in user_info and user_info[username] == password:
    print("Login Successful")
    return True
  else:
    print("Login Failed")
    return False

while True:
  print("1. Sign Up\n2. Login\n3. Exit")
  choice = int(input("Enter choice: "))
  if choice == 1:
    username = input("Enter username: ")
    password = input("Enter password: ")
    signUp(username, password)

  elif choice == 2:
    username = input("Enter username: ")
    password = input("Enter password: ")
    login(username, password)
  elif choice == 3:
    break
  else:
    print("Invalid choice")
```

```python
#Counting Elements with a Dictionary
words = input("Enter the words with spaces: ").split()
word_count = {}
for word in words:
  if word in word_count:
    word_count[word] += 1
  else:
    word_count[word] = 1
print(word_count)
```

```python
#Temperature Converter
def convert(val, choice):
  if choice == 'CtoF':
    return (val * 9/5) + 32
  elif choice == 'FtoC':
    return (val - 32) * 5/9
  else:
    print("Invalid choice")
    return 0
convert(21,'CtoF')
```
